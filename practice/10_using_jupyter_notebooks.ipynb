# Pandas in Jupyter Notebooks

**What is a jupyter notebook?**

Notice that the file extension of this file is .ipynb instead of just .py, meaning we are looking at a Jupyter notebook. A notebook is a document made of **cells**, which can either be markdown cells, or code cells. Markdown cells hold notes, while code cells hold Python. You can mix and match and arrance different cells however you like to have explanations right next to small chunks of code. You run code one cell at a time, see results immediately, and iterate quickly. This is great for quick data exploration and reporting.


## Viewing Pandas dataframes in jupyter notebooks
We'll load in the mock sleeping data below. Notice how jupyter notebooks lets you view the table in a nice way. Also try opening up the "Data Wrangler" view.
Jupyter notebooks (.ipynb) are primarily used to analyze data, while .py are used to run programs and scripts. 

```python
import pandas as pd

df = pd.read_excel("../mock_sleeping_happiness_data.xlsx")
df # This will show up as a nice table. You don't need to print, and in fact it is better if you don't
```

## Looking at a subset of data, visualizing data
You can make lots of separate dataframes. Look at the "Jupyter Variables" section. You can examine each dataframe (or any other variable you make) without having your code actively running.

```python
avg_happiness = (
    df.groupby("gender")["happiness_rating"].mean().sort_values(ascending=False)
)
avg_happiness
```

```python
import subprocess, sys
try:
    import matplotlib.pyplot as plt
except ImportError:
    print("You don't have matplotlib installed yet, installing now...")
    subprocess.check_call([sys.executable, "-m", "pip", "install", "matplotlib"])
    import matplotlib.pyplot as plt

ax = avg_happiness.plot(kind="bar")  # uses matplotlib under the hood
ax.set_title("Average Happiness by Gender")
ax.set_xlabel("Gender")
ax.set_ylabel("Average Happiness")
plt.show()
```

## Why use notebooks with pandas?
- **Iterative exploration**: Change a cell, re-run, and immediately see 
  results.
- **Blend code and narrative**: Mix explanations (Markdown) with code outputs.
- **Shareable**: One file shows the steps, the reasoning, and the results.
- **Great for quick experiments**: Try an idea without building a full script.
